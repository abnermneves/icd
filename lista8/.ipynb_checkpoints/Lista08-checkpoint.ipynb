```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support

from sklearn.neighbors import KNeighborsClassifier

# Um fator importante é que o SKLearn não cria conjuntos de validação para você. Você tem algumas abordagens,
# uma é realizar um novo split no treino. Outra é fazer uso de classificadores com CV no fim.
# Tipo LogisticRegressionCV (ver na API). Por fim, você pode fazer uso da classe GridSearchCV.
# Leia a documentação da mesma.
from sklearn.model_selection import train_test_split
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
white = pd.read_csv('winequality-white.csv', sep=';')
red = pd.read_csv('winequality-red.csv', sep=';')
red.head()
```

```python
# juntando as duas tabelas e one hot encoding pra cor do vinho

white['white'] = pd.Series(np.ones(len(white))).astype('int')
white['red'] = pd.Series(np.zeros(len(white))).astype('int')

red['white'] = pd.Series(np.zeros(len(red))).astype('int')
red['red'] = pd.Series(np.ones(len(red))).astype('int')

df = white.append(red, ignore_index=True)
df
```

```python
# separando X e y

y = df['quality'].copy()
X = df.copy()
X = X.drop('quality', axis=1)
```

```python
# separando e normalizando train e test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

numerics = X_train.columns[X_train.columns != 'white']
numerics = numerics[numerics != 'red']

train_mean = X_train.loc[:, numerics].mean()
train_std = X_train.loc[:, numerics].std(ddof=1)

Z_train = X_train.copy()
Z_train[numerics] = X_train[numerics] - train_mean
Z_train[numerics] = Z_train[numerics] / train_std

Z_test = X_test.copy()
Z_test[numerics] = X_test[numerics] - train_mean
Z_test[numerics] = Z_test[numerics] / train_std
Z_test
```

<h1>KNN</h1>

```python
skf = StratifiedKFold(n_splits=5)
fold = 1
for train_index, val_index in skf.split(Z_train, y_train):
    print("Fold: ", fold)
    for k in range(1, 16, 2):
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(Z_train.values[train_index], y_train.values[train_index])
        
        y_val = y_train.values[val_index]
        y_prev = model.predict(Z_train.values[val_index])
        
        metricas = precision_recall_fscore_support(y_val, y_prev, average='micro')
        print("K = ", k, "Métricas: ", metricas[:3]) # precisão, revocação e F1

    fold += 1
    print()
    print()
    
```

O melhor parâmetro parece ser K = 1, pois em todos os casos teve as melhores métricas. Vamos confirmar com o GridSearchCV:

```python
params = {'n_neighbors': np.arange(1, 16, 2)}
gs = GridSearchCV(model, params, cv=5)
gs.fit(Z_train, y_train)
best_k = gs.best_params_['n_neighbors']
best_k
```

Agora testamos o modelo com o subconjunto de teste

```python
model = KNeighborsClassifier(n_neighbors=best_k)
model.fit(X_train, y_train)

prev = model.predict(X_test)

metricas = precision_recall_fscore_support(y_test, prev, average='micro')

confusao = confusion_matrix(y_test, prev)

print("Precisão: ", metricas[0])
print("Revocação: ", metricas[1])
print("F1: ", metricas[2])
print("Matriz de confusão: ", '\n', confusao)
print("Classes de qualidade: ", y.unique())
```

<h1>Regressão logística</h1>

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
reg = LogisticRegressionCV(cv=5, random_state=42)
reg = reg.fit(Z_train, y_train)

y_val = y_test.values
prev = reg.predict(Z_test)

metricas = precision_recall_fscore_support(y_test, prev, average='micro')

confusao = confusion_matrix(y_test, prev)
print("Precisão: ", metricas[0])
print("Revocação: ", metricas[1])
print("F1: ", metricas[2])
print("Matriz de confusão: ", '\n', confusao)
print("Classes de qualidade: ", y.unique())
```
